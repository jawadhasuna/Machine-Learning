## TASK 1 EDA CHURN DATA SET

```python
import pandas as pd
df = pd.read_csv("churn.csv")
df=df.transpose()
print('head:',df.head())
print('tail:',df.tail())
df=df.transpose()
```

```python
print('shape:', df.shape)
print('columns:', df.columns)
print('total charges sum:',df['TotalCharges'].sum())
print('monthly charges mean:',df['MonthlyCharges'].mean())
print('monthly charges sum:',df['MonthlyCharges'].sum())
df.drop('customerID', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
print('tenure mean:',df['tenure'].mean())
print('internet service:',df['InternetService'].nunique())
print('total charges 0-10:\n',df.loc[0:10,'TotalCharges'])
print('10 random:\n',df.sample(10))
```

```python
tenure=df['tenure'].values
print('tenure array:\n',tenure)
print('max:\n',tenure.max())
print('min:\n',tenure.min())
print('mean:\n',tenure.mean().round(6))
```
