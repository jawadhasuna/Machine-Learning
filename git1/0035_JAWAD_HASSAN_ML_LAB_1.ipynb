```python
#Task 2
bookinvent=[]
def addbook():
  t=input('title:')
  a=input('author:')
  y=int(input('year:'))
  i=int(input('isbn:'))
  q=int(input('quantity:'))
  bookinvent.append((t,a,y,i,q))
def searchbook():
  ii=int(input('isbn:'))
  for book in bookinvent:
      if ii==book[3]:
       print(book)

def displayinventory():
  for books in bookinvent:
    print(books,end='\n')

def sell():
  isbn=int(input('isbn:'))
  quan=int(input('quan:'))
  for i in range(len(bookinvent)):
    if isbn==bookinvent[i][3]:
      if quan<=bookinvent[i][4]:
        temp=list(bookinvent[i])
        temp[4]-=quan
        bookinvent[i]=tuple(temp)
        print('sold successfully')
      else:
        print('out of stock')
      return
  print('not available')


addbook()
addbook()
displayinventory()
searchbook()
sell()
displayinventory()



```
